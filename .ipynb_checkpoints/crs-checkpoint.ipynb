```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing dataset
df=pd.read_csv(r"C:\Users\saran\Desktop\CRS\data\courses.csv")
```

```python
df.head(5)
```

```python
df.info()
```

```python
# Replacing ';' --> '|'

cols_to_fix = ['skills_covered','topics','prerequisites','career_tags', 'learning_outcomes']
for i in cols_to_fix:
    df[i]=df[i].str.replace(';','|', regex= 'False')
```

```python
difficulty_map = {'Beginner':1,'Intermediate' :2 , 'Advanced' : 3}
df['difficulty_num'] = df['difficulty_level'].map(difficulty_map)
```

```python
df.head()
```

```python
list_cols= ['skills_covered','topics','prerequisites','career_tags', 'learning_outcomes']
for i in list_cols:
    df[i]=df[i].str.split('|')
```

```python
# creating one single(smart) feature
df['combined_features']=df['domain'] +' ' + df['difficulty_level'] + df['skills_covered'].apply(lambda x: ' '.join(x))+ df['topics'].apply(lambda x: ' '.join(x)) + df['career_tags'].apply(lambda x: ' '.join(x))
df['combined_features']
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer( stop_words = 'english', ngram_range= (1,2) , max_features = 600)
tfidf_matrix = tfidf.fit_transform(df['combined_features'])
```

```python
tfidf_matrix.shape
```

```python
# What models learned
feature_names= tfidf.get_feature_names_out()

tfidf_df= pd.DataFrame(
    tfidf_matrix.toarray(), columns = feature_names)
tfidf_df.head()
```

```python
tfidf_df.iloc[0].sort_values(ascending=False).head(10)
```

```python
# cosine similarity 
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix)
```

```python
cosine_sim.shape
```

```python
cosine_sim[2]
```

```python
# recommendor system
def recommendor_courses(course_index, top_n=5):
    scores = list(enumerate(cosine_sim[course_index]))
    scores = sorted(scores, key=lambda x: x[1], reverse= True)

    scores =scores[1:top_n+1]
    indices=[i[0] for i in scores]
    return df.iloc[indices]
```

```python
recommendor_courses(2,5)
```
